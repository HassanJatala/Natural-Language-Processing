```python
import nltk
from nltk.stem import PorterStemmer
```

```python
text = "I AM Loving the NLP Class!!! But sometimes, it Feels very CONFUSING !!!"
print("Original Text:")
print(text)

text = text.lower()
print("\nAfter Lowercasing:")
print(text)
```

```python
ps = PorterStemmer()
words = text.split()
stemmed = [ps.stem(w) for w in words if w.isalpha()]
print("\nStemming Result:")
print(stemmed)
```
