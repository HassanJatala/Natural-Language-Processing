```python
from gensim.models import Word2Vec
```

```python
sentences = [
    ["i", "love", "machine", "learning"],
    ["machine", "learning", "is", "fun"],
    ["i", "love", "ai"]
]
```

```python
model = Word2Vec(
    sentences,
    vector_size=50,
    window=5,
    min_count=1,
    workers=4
)
```

```python
vector = model.wv["machine"]
print("Vector size:", len(vector))
print(vector)
```

```python
print("\nSimilar to 'machine':")
print(model.wv.most_similar("machine"))
```
